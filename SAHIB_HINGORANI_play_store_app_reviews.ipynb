# ***DATA CLEANING***

As we can see from the above playstore data, we will have to clean some of the data to make our analysis much more accurate and user friendly.


 Shift all the entries by one for row no. 10475 
 app name- “Life Made Wi-Fi Touchscreen Photo Frame”

```python
play_store_data_df.iloc[10472] = play_store_data_df.iloc[10472].shift(1)
play_store_data_df.at[10472, "App"] = play_store_data_df.iloc[10472]["Category"]
play_store_data_df.at[10472, "Category"] = "UNKNOWN"
play_store_data_df.iloc[10472]
```

```python
# convert 'Reviews' object into Float



R= [i for i in play_store_data_df['Reviews']]



play_store_data_df['Reviews'] = play_store_data_df['Reviews'].astype(float)
```

```python
# Clean the 'Size' convert 'Size' object into Float
# While analyzing the data in pivot table found a value with 1000+



S = [i for i in play_store_data_df['Size']]

def clean_sizes (sizes):

  Cleaned_data=[]

  for s in sizes:
    s = str(s)
    
    if 'M' in s:
      s = s.replace('M' , '')
      s=float(s)
      
      


    elif 'k' in s:
      s= s.replace('k', '')
      s=float(s)
      s= s/1024                   

    elif 'Varies with device' in s:
      s=float(0)
      

    Cleaned_data.append(s)

  return Cleaned_data


play_store_data_df['Size'] = clean_sizes(S)
play_store_data_df['Size']= play_store_data_df['Size'].astype(float)
```

We have converted all the diffrent vlaues under one value i.e k into M,M is already M
So the cleaned data will have apps with size denoted with 'M'

```python
# clean Installs and change datatype from object to float

I = [i for i in play_store_data_df['Installs']]

def clean_installs(install):
  Cleaned_data=[]
  for x in install:

    
      
    if ',' in x:
      x=x.replace(',','')

    elif '+' in x:

      x=x.replace('+' , '')

    
      

    Cleaned_data.append(x)

  return Cleaned_data

play_store_data_df['Installs']= clean_installs(I)
play_store_data_df['Installs']= play_store_data_df['Installs'].astype(float)

  
```

```python
# Clean 'Price' and convert datatype object to Float

P= [i for i in play_store_data_df['Price']]

def clean_prices(price):
  Cleaned_data=[]
  
  for p in price:
    
    if '$' in p:
      p=p.replace('$','')


    

    Cleaned_data.append(p)

  return Cleaned_data



play_store_data_df['Price']= clean_prices(P)
play_store_data_df['Price']= play_store_data_df['Price'].astype(float)

```

# **ANALYSIS on Play Store Data** 

Now as we are done with data preparation and data cleaning, we can go ahead with analysis of the data.

No-1).   Categories the app developer(s) prefer.

```python
Categories = play_store_data_df['Category'].value_counts()
Categories.head()
```

```python
sns.set_style('ticks')
plt.figure(figsize=(12,7))
sns.countplot(y='Category', data=play_store_data_df)
plt.title('No. of apps per category')
plt.show()
```

As we saw from the above data and graph, we can conclude that application developers prefer to build the app under Family category followed by Games and Tools

No.2). Content rating given by PEGI(App Content Rating Statistics for Google Play)

      -Target audience for app developers.
      

```python
import plotly.express as px

x = play_store_data_df.groupby(['Content Rating'],as_index=False)['App'].count()
px.pie(x,values='App',names='Content Rating',title='Target audience for App developers').show()
```

With the help of above analysis we can see that audience that most of the app developers target is EVERYONE that are meant to be for all age groups.

Which is followed by Teen and Mature 17+

Further we can check the distribution of categories across diffrent kinds of content rating

```python
TA = play_store_data_df.groupby(['Content Rating','Category'],as_index=False)['App'].count()

px.sunburst(TA, values='App', path=['Content Rating','Category'], title='Categories distribution over Target audience',color='Category')
```

Apps distribution between content rating on a sunburst chart, and interestingly apps percentage were different for all.

**EVERYONE** - Family apps were leading followed by Tools and Games.

**TEEN** - Games and then Family.

**MATURE 17+** - Dating apps are leading their way followed by Games

**EVERYONE 10+** - Family and Games are leading to others.

No.3)  Now we will check types of apps in terms of their value, weather they are free of cost to users or users are been charged for using the apps

```python
plt.figure(figsize=(8, 4))
sns.countplot(play_store_data_df['Type'])
plt.ylabel('Number of Apps')
plt.title('Type Distribution')
plt.show()
```

Majority of the apps are free to users which I have plotted in a countplot

Also in a scatterplot below just to show it in a different way.

Throuh the above sccater plot, we can see that there are ample number of free apps present in the play store market and whereas there are only limited number of apps which are paid

No.4) Size of the apps per categories



```python
ncount=1
plt.figure(figsize=(20,30))
for i in np.unique(play_store_data_df.Category.values):
    plt.subplot(11,3,ncount)
    plt.hist(play_store_data_df[play_store_data_df.Category == i].Size.values)
    plt.title(i)
    plt.ticklabel_format(useOffset=False, style='sci')
    plt.xlabel("Count")
    plt.ylabel("Size per mb")
    ncount = ncount + 1
plt.tight_layout()
```

With the above analysis we can check that Category Family and Tools has the highest amount of sizes as compared to the other categories.

No.5)  Now we are going to check the correlation between Rating, Reviews, Installs and Price.

From the above Heatmap, we can see that the relationship between the factors mentioned above.
From the above data it is clear that there is strong positive correlation between reviews and installs.
